```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('files/titanic.csv')
```

```python
data.head()
```

```python
### Count the male and female with visualization
```

```python
plt.figure(figsize=(12,8))

sns.countplot(data=data,x='sex')
plt.show()
```

```python
#spliting data
```

```python
plt.figure(figsize=(12,8))
sns.set(palette='rainbow')
sns.countplot(data=data,y='sex',hue='pclass')
plt.show()
```

```python
###################
#
#Categorical Plots all male and female types of passenger class and their age
```

```python
plt.figure(figsize=(12,8))
sns.stripplot(data=data,x='sex',y='age',jitter=True,hue=data['pclass'],dodge=True)
plt.show()
#scatter
```

```python
#scatter using the survived column
```

```python
plt.figure(figsize=(12,8))
sns.stripplot(data=data,x='sex',y='age',jitter=True,hue=data['survived'],dodge=True)
plt.show()
```

```python
####
# swarmplot
# non overlapping point
# better view of scatterplot
```

```python
plt.figure(figsize=(12,8))
sns.swarmplot(data=data,x='sex',y='age',hue=data['pclass'],dodge=True)
plt.show()
```

```python
###
# violinplot
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=data,x='sex',y='age',hue=data['pclass'],dodge=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=data,x='sex',y='age',hue=data['sex'],dodge=True,split=True)
sns.swarmplot(data=data,x='sex',y='age',hue=data['survived'],dodge=True)
plt.show()
```

```python
###########
```

```python
#Barplot
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=data,x='pclass',y='age',hue=data['sex'])
plt.show()
```

```python
## pointplot
```

```python
plt.figure(figsize=(12,8))
sns.pointplot(data=data,x='pclass',y='age',hue=data['sex'],dodge=True)
plt.show()
```
