```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(palette='rainbow')
```

```python
data = pd.read_csv('files/titanic.csv')
```

```python
data.head()
```

```python
#######
# JOIN PLOT
```

```python
plt.figure(figsize=(16,10))
sns.jointplot(data=data,x='age',y='fare',height=8,kind='scatter')
plt.show()
```

```python
plt.figure(figsize=(14,10))
sns.jointplot(data=data,x='age',y='fare',height=8,kind='hex')
plt.show()
```

```python
#kind= 'kde'
```

```python
#Regression plot
```

```python
sns.jointplot(data=data,x='age',y='fare',height=8,kind='reg')
plt.show()
```

```python
####
## using LMPLOT
```

```python
sns.lmplot(data=data,x='age',y='fare',aspect=1,height=8,hue='sex')
plt.show()
```

```python
sns.lmplot(data=data,x='age',y='fare',aspect=1,height=8,col='sex')
plt.show()
```

```python
#chance to survive
```

```python
sns.lmplot(data=data,x='age',y='survived',aspect=1,height=8,logistic=True,col='sex')
plt.show()
```

```python
#using pclass
```

```python
sns.lmplot(data=data,x='age',y='survived',aspect=1,height=8,col='pclass',logistic=True)
plt.show()
```
